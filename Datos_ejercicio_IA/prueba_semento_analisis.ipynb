```python
import pandas as pd
```

```python
df = pd.read_csv('D:/Inteligencia artifical/Proyecto_f_TalentT/CPN50_cement_composition_strength.csv', sep=';', encoding='latin1')
df.head() 
```

```python
df.count()
```

```python
#1 Verificar valores nulos
print("\n Varlores nulos por columna: ")
print(df.isnull().sum())
```

```python
valores_unicos =df['r1_iram1622'].dropna().unique() #dropna elimina datos faltantes
print("\n iram únicos: ")
print(valores_unicos)
```

```python
# Convertir múltiples columnas a numéricas y filtrar solo filas con valores positivos
columnas = ['r1_iram1622', 'r2_iram1622', 'r3_iram1622', 'r7_iram1622', 'r28_iram1622', 'r91_iram1622']  # Lista de columnas a convertir

# Filtrar las columnas que no están completamente vacías
columnas_con_datos = [col for col in columnas if df[col].notnull().any()]

# Convertir las columnas filtradas a numéricas
for col in columnas_con_datos:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Filtrar filas donde todas las columnas con datos sean mayores a 0
df_filtrado = df[columnas_con_datos].dropna()  # Eliminar filas con NaN en las columnas con datos
df_filtrado = df_filtrado[(df_filtrado > 0).all(axis=1)]  # Mantener solo filas con valores positivos

# Recuperar el DataFrame original filtrado
df_filtrado = df.loc[df_filtrado.index]

df_filtrado
```

```python
print("\n Varlores nulos por columna: ")
print(df_filtrado.isnull().sum())
```

```python
# Mostrar las filas donde 'g75µ' o 'na2o' tengan valores nulos
filas_nulas = df_filtrado[df_filtrado[['g75µ', 'na2o']].isnull().any(axis=1)]

# Mostrar las filas con valores nulos
filas_nulas
```

```python
# Eliminar las filas con valores nulos en 'g75µ' o 'na2o' del DataFrame original
df_filtrado = df_filtrado.drop(filas_nulas.index)

# Mostrar el DataFrame filtrado sin esas filas
df_filtrado
```

```python
print("\n Varlores nulos por columna: ")
print(df_filtrado.isnull().sum())
```

```python
df_filtrado.count()
```

```python
# Contar los valores duplicados en cada columna de df_filtrado
for col in df_filtrado.columns:
    duplicated_counts = df_filtrado[col].value_counts()  # Contar los valores en la columna
    duplicated_counts = duplicated_counts[duplicated_counts > 1]  # Filtrar los que aparecen más de una vez
    
    if not duplicated_counts.empty:  # Solo mostrar si hay duplicados
        print(f"Valores duplicados en la columna '{col}':")
        print(duplicated_counts)
        print("\n" + "-"*40 + "\n")

```
